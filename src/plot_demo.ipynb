```julia
using Plots
```

```julia
x = collect(0:0.1:2*pi);
```

```julia
y = sin.(x);
```

```julia
plot(x,y)
```

```julia
plot(x,y, color = :red, title="first example")
```

```julia
plot(x,y, linestyle = :dot, arrow = :arrow, linealpha = 0.5, linewidth = 4, linecolor = :red)
```

```julia
plot(x, y, markershape = :hexagon,
    markersize = 4,
    markeralpha = 0.6,
    markercolor = :green,
    markerstrokewidth = 2,
    markerstrokealpha = 0.2,
    markerstrokecolor = :red,
    markerstrokestyle = :dot)
```

```julia
plot(x,y, 
    xlabel = "my label",
    xlims = (0,2*pi),
    xticks = 0:0.5:2*pi,
    xflip = true,
    xtickfont = font(4, "Courier")
)
```

```julia
n = collect(1:10);
```

```julia
y = 0.2.^n;
```

```julia
plot(n, y)
```
